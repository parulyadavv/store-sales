```python
from sklearn.preprocessing import LabelEncoder
```

```python
def categTheDate(feature,start):
    le = LabelEncoder()
    feature = feature.astype(str)
    encoded_feature = le.fit_transform(feature)
    return encoded_feature+start
```

```python
train_data.columns
```

```python
columns_to_encode = [ 'family', 'city',
       'state', 'type_x', 'type_y', 'locale', 'locale_name',
       'description']

for column in columns_to_encode:
    train_data[column] = categTheDate(feature=train_data[column],start=1)
```
